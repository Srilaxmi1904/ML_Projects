## SPLITTINF THE DATA AS TRAINING AND TESTING DATA

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(data['Text'],data['Language'],test_size=0.30,random_state=42)
x_train
y_train
```

## USING COUNT VECTORIZER FOR TEXT ANALYSIS

## AND IMPLEMENTING THE ALGORITHM

```python
from sklearn.feature_extraction.text import CountVectorizer
count=CountVectorizer()
count.fit(x_train)
x_train_count=count.fit_transform(x_train)


from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(x_train_count,y_train)

text=['എങ്ങിനെ ഇരിക്കുന്നു']

text_count=count.transform(text)
model.predict(text_count)
```
