## READING THE DATA

```python
import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_csv('/content/voice.csv')
data
```

```python
data.head()
```

```python
data.isna().sum()
```

```python
data.columns

#Q25  Q75  IQR  skew  sp.ent  sfm  centroid  
```

## DROPPING UNNECESSARY COLUMNS:

```python
data=data.drop(['Q25','Q75','IQR','skew','sp.ent','sfm','centroid'],axis=1)
```

```python
data.head()
```

## SEPARATING X(INPUT) AND Y(OUTPUT) VALUES:

```python
x=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
```

## SPLITTING THE DATA AS TRAINING AND TESTING DATA

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

## NORMALIZATION OF THE DATA

```python
from sklearn.preprocessing import StandardScaler
scal=StandardScaler()
scal.fit(x_train)
x_train=scal.transform(x_train)
x_test=scal.transform(x_test)
x_test
```

## IMPLEMENTING SVM ALGORITHM

```python
from sklearn.svm import SVC
model=SVC()
model.fit(x_train,y_train)

pred=model.predict(x_test)
pred

print(model.predict([['0.151228','0.072111','0.158011','4.177296','0.083878','0.088965','0.017798','0.250000','0.201497','0.007812','0.562500','0.554688','0.247119']]))
```

## MODEL PERFORMANCE EVALUATION

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix,ConfusionMatrixDisplay

con=confusion_matrix(y_test,pred)
con

score=accuracy_score(y_test,pred)
score

rep=classification_report(y_test,pred)
print(rep)
```

```python
con  #confusion matrix
label=['male','female']
display=ConfusionMatrixDisplay(con,display_labels=label)
display.plot()
```
