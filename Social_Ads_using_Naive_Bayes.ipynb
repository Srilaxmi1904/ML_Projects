## READING THE DATA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data=pd.read_csv('/content/Social_Network_Ads.csv')
data
```

```python
data.head()
```

```python
data.isna().sum()
```

## LABEL ENCODING (to convert text labels to numeric form so that machine can read it easily)

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data['Gender']=le.fit_transform(data['Gender'])
data
```

## DELETING UNNECESSARY COLUMNS FROM THE DATA

```python
data=data.drop(['User ID'],axis=1)
data
```

```python
count=data['Age'].value_counts().sort_values().rename_axis('Age').reset_index(name='counts')
count
```

## SEPARATING X(INPUT) AND Y(OUTPUT) VALUES FROM THE DATA

```python
x=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
```

## SPLITTING THE DATA AS TRAINING AND TESTING DATA

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

## IMPLEMENTING THE ALGORITHM (GaussianNB as the data is numerical)

```python
from sklearn.naive_bayes import GaussianNB
gau=GaussianNB()
gau.fit(x_train,y_train)

prediction=gau.predict(x_test)
prediction

print(gau.predict([[1,45,22000]]))
```

## EVALUATING THE PERFORMANCE OF THE MODEL

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
res=confusion_matrix(y_test,prediction)
res

score=accuracy_score(y_test,prediction)
score

classes=classification_report(y_test,prediction)
print(classes)
```
