```python
#Convert string columns to numerical columns

#Here we will use GET DUMMY (comes under Pandas)
#very similiar to ONE HOT ENCODER.

#we can reduce 1 feature of x in get dummies.

#ex:  in one hot encoding:
#Apple      #Orange      Grape      Mango
# 1            0           0           0       Apple = 1000
# 0            1           0           0       Orange= 0100
# 0            0           1           0       Grape=  0010
# 0            0           0           1       Mango=  0001
#so if there are 10 fruits ,there will be 10 features


#In GET DUMMIES we can reduce the feature as 9: How?
# Apple column will be removed.
##Orange      Grape      Mango
# 0           0           0       Apple = 000
# 1           0           0       Orange= 100
# 0           1           0       Grape=  010
# 0           0           1       Mango=  001

```

## CONVERTING STRING COLUMNS TO NUMERICAL VALUES USING GET-DUMMIES

```python
dummy=pd.get_dummies(data[['Location','Transmission','Fuel_Type','Owner_Type']],drop_first=True)
dummy
```

## COMBINING OLD DATA + DUMMY DATA USING CONCAT FUNCTION

```python
dd1=pd.concat([data,dummy],axis=1)
dd1

#first we had only 14 columns now it is 32 columns
```

## DROPPING UNNECESSARY COLUMNS

```python
dd1.drop(['Location','Fuel_Type','Transmission','Owner_Type','Unnamed: 0','Name','New_Price'],axis=1,inplace=True)
```

```python
dd1
```

## REMOVING STRING VALUES AMONG NUMBERS FROM THE COLUMNS

```python
dd1['Mileage']=dd1['Mileage'].str.replace('km/kg','')
dd1['Mileage']=dd1['Mileage'].str.replace('kmpl','')
dd1['Engine']=dd1['Engine'].str.replace('CC','')
dd1['Power']=dd1['Power'].str.replace('bhp','')
dd1['Mileage']=dd1['Mileage'].str.replace('null','0')
dd1['Power']=dd1['Power'].str.replace('null','0')
dd1['Engine']=dd1['Engine'].str.replace('null','0')
dd1.head()
```

```python
dd1.dtypes
```

## CONVERTING OBJECTS TO INT/FLOAT

```python
dd1['Engine']=dd1['Engine'].astype(float)
dd1['Power']=dd1['Power'].astype(float)
dd1['Mileage']=dd1['Mileage'].astype(float)
dd1.dtypes
```

## CONVERTING COLUMNS WITH 0 VALUES TO 'NaN'

```python
#in MILEAGE it is impossible for a car to have mileage 0. In datadummy we have them. So I have to replace it.
#converting 0 to NaN
dd1.loc[dd1.Engine==0,'Engine']=np.NaN
dd1.loc[dd1.Power==0,'Power']=np.NaN
dd1.loc[dd1.Mileage==0,'Mileage']=np.NaN
dd1
```

```python
dd1.isna().sum()
```

## FILLING THE VACCUME COLUMNS

```python
dd1['Engine']=dd1['Engine'].fillna(dd1['Engine'].mean())
dd1['Mileage']=dd1['Mileage'].fillna(dd1['Mileage'].mean())
dd1['Power']=dd1['Power'].fillna(dd1['Power'].mean())
dd1['Seats']=dd1['Seats'].fillna(dd1['Seats'].mode() [0])
dd1.isna().sum()
```

## REARRANGING THE DATA INDEX

```python
oldcol=dd1.columns.values
oldcol
newcol=['Year', 'Kilometers_Driven', 'Mileage', 'Engine', 'Power', 'Seats', 'Location_Bangalore', 'Location_Chennai',
       'Location_Coimbatore', 'Location_Delhi', 'Location_Hyderabad',
       'Location_Jaipur', 'Location_Kochi', 'Location_Kolkata',
       'Location_Mumbai', 'Location_Pune', 'Fuel_Type_Diesel',
       'Fuel_Type_Electric', 'Fuel_Type_LPG', 'Fuel_Type_Petrol',
       'Transmission_Manual', 'Owner_Type_Fourth & Above',
       'Owner_Type_Second', 'Owner_Type_Third','Price']
dd1=dd1.reindex(columns=newcol)
dd1.head()
```

```python
dd1.drop(['Fuel_Type_Electric'],axis=1,inplace=True)
```

```python
dd1
```
